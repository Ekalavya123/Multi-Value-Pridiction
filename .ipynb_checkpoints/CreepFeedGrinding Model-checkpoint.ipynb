```python
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow.keras as keras
import matplotlib.pyplot as plt
import pandas as p
data=p.read_excel("temp_data1.xlsx")
time,temp=data["Time [s]"],data["Maximum [°C]"]
#print(time,temp)
X_train, X_test, y_train, y_test = train_test_split(time, temp, test_size=0.3)
print(y_test)
#print(X_train, X_test, y_train, y_test)

model = keras.Sequential([
    # input layer
    keras.layers.Flatten(input_dim = 1),
    # 2nd dense layer
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(16, activation='relu'),
    
    # output layer
    keras.layers.Dense(1, activation='linear')
])

# compile model
optimiser = keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimiser,
              loss='MeanSquaredError',
              metrics=['accuracy'])

model.summary()
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=1, epochs=50)
```

```python
# print(temp)
# print(X_train, X_test, y_train, y_test)
# print(time)
t= np.linspace(0,3,50)
print(model.predict([1]))
print(t)
pridiction=model.predict(t) 
print(pridiction)
plt.plot(t,pridiction,'o')
plt.plot(t,pridiction)
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow.keras as keras
import matplotlib.pyplot as plt
import pandas as p
heat_data=p.read_excel("heat-flux1.xlsx")
time,heat=heat_data["Time [s]"],heat_data["Maximum [W/m²]"]
#print(time,temp)
X_train, X_test, y_train, y_test = train_test_split(time,heat, test_size=0.3)
#print(X_train, X_test, y_train, y_test)
#print(y_test)

model = keras.Sequential([
    # input layer
    keras.layers.Flatten(input_dim = 1),
    # 2nd dense layer
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(16, activation='relu'),
    
    # output layer
    keras.layers.Dense(1, activation='linear')
])

# compile model
optimiser = keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimiser,
              loss='MeanSquaredError',
              metrics=['accuracy'])

model.summary()
model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=1, epochs=50)
```

```python
t=[]
pred_temp,pred_heat,pred_time=[],[],[]
i=0
for _ in range(50):
    i+=(3/50)
    t.append([i,i])
    pred_time.append(i)
print(t)

pridiction=model.predict(t) 
for i in range(len(t)):
    pred_temp.append(pridiction[i][0])
    pred_heat.append(pridiction[i][1])
print(pridiction)
plt.plot(pred_time,pred_temp,'o')
plt.plot(pred_time,pred_temp)
plt.show()
```
