```python
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow.keras as keras
import matplotlib.pyplot as plt
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
def make_model(file,column,scale):
    data=p.read_excel(file)
    time,oup=data["Time [s]"],data[column]
    for i in range(len(oup)):
        if(scale!=1): oup[i]=oup[i]/scale
    #print(time,oup)
    X_train, X_test, y_train, y_test = train_test_split(time,oup, test_size=0.3)
    #print(X_train, X_test, y_train, y_test)
    model = keras.Sequential([
        # input layer
        keras.layers.Flatten(input_dim = 1),
        
        # hidden dense layers
        keras.layers.Dense(128, activation='relu'),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.Dense(16, activation='relu'),

        # output layer
        keras.layers.Dense(1, activation='linear')
    ])

    # compile model
    optimiser = keras.optimizers.Adam(learning_rate=0.01)
    model.compile(optimizer=optimiser,
                  loss='MeanSquaredError',
                  metrics=['mean_squared_error'])

    model.summary()
    model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=10, epochs=100)
    return model
m1_heat_model=make_model("heat_flux1.xlsx","Maximum [W/m²]",1000)
m1_temp_model=make_model("temp_data1.xlsx","Maximum [°C]",1)
m2_heat_model=make_model("heat_flux2.xlsx","Maximum [W/m²]",1000)
m2_temp_model=make_model("temp_data2.xlsx","Maximum [°C]",1)
```

```python
t=np.linspace(0,3,50)
m1_temp_pred=m1_temp_model.predict(t)
m2_temp_pred=m2_temp_model.predict(t)
plt.plot(t,m1_temp_pred,label="partially stabilized zirconia")
plt.plot(t,m2_temp_pred,label="AISI 51200")
plt.legend()
plt.title("Temperature Distribution")
plt.xlabel("Time [s]")
plt.ylabel("Temp[°C]")
plt.show()
m1_heat_pred=m1_heat_model.predict(t)
m2_heat_pred=m2_heat_model.predict(t)
plt.plot(t,m1_heat_pred,label="partially stabilized zirconia")
plt.plot(t,m2_heat_pred,label="AISI 51200")
plt.legend()
plt.title("Heat Flux Distribution")
plt.xlabel("Time [s]")
plt.ylabel("Heat Flux[KW/m²]")
plt.show()
```
